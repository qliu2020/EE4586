```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt 
from sklearn.model_selection import StratifiedKFold
from sklearn.base import clone
from sklearn.preprocessing import StandardScaler
import seaborn as sns
from scipy.stats import pearsonr

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn_rvm import EMRVC
from sklearn import svm
```

```python
data = pd.read_csv('winequality-red.csv')
features = ['fixed acidity','volatile acidity','citric acid','residual sugar','chlorides',
            'free sulfur dioxide','total sulfur dioxide','density','pH','sulphates','alcohol']
corr=data[features].corr()
sns.heatmap(corr,xticklabels=corr.columns.values,yticklabels=corr.columns.values)
plt.show()

F=True
for i in range(len(features)):
    for j in range(i+1,len(features)):
        if abs(corr.iloc[i,j])>=0.6:
            print('The feature %s and feature %s is highly correlated with coefficient %f'%(features[i],features[j],corr.iloc[i,j]))
            F=False
if F:
    print('No highly correlated features')
X = data.iloc[:,0:11].values
y = data.iloc[:,11].values
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
scaler.fit(X_test)
X_test = scaler.transform(X_test)

clf = RandomForestClassifier(n_estimators = 60)
#clf = EMRVC(kernel="rbf")
#clf = svm.SVC()
clf.fit(X_train, y_train)
print(clf.score(X_test, y_test))
```

```python
skfolds = StratifiedKFold(n_splits=6)
clf = RandomForestClassifier(n_estimators = 60)
#clf = svm.SVC()
score=[]
scaler = StandardScaler()
for train_index, test_index in skfolds.split(X, y):
    X_train_folds = X[train_index]
    scaler.fit(X_train_folds)
    X_train_folds = scaler.transform(X_train_folds)
    y_train_folds = y[train_index]
    
    X_test_fold = X[test_index]
    scaler.fit(X_test_fold)
    X_test_fold = scaler.transform(X_test_fold)
    y_test_fold = y[test_index]
    
    clone_clf = clone(clf)    
    clone_clf.fit(X_train_folds, y_train_folds)
    score.append(clone_clf.score(X_test_fold, y_test_fold))
    
print(np.mean(score))
```
